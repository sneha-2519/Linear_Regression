# Simple Linear Regression Model

## Step 1 : import necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Step 2 : Load the dataset

```python
df = pd.read_csv(r"E:\Sneha\machine learning datasets\regression model\Salary_Data.csv")
df
```

## Step 3 : Data Analysis

```python
df.describe()
```

```python
plt.title("Salary distribution")
sns.distplot(df['Salary'])
plt.show()
```

## relationship between Salary and Experience

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel("Salary")
plt.ylabel("Year of experiance")
plt.show()
```

### It is clearly visible now, our data varies linearly. That means, that an individual receives more Salary as they gain Experience.

## Step 4 : Split the dataset into dependent/independent variables

```python
# Splitting variables

X = df.iloc[:, :1]  # independent
y = df.iloc[:, 1:]  # dependent
```

## Step 5 : Split data into Train/Test sets (training (80%) and test (20%) sets )

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Step 6 : Train the regression model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

### Accuracy with testing data

```python
lr.score(X_test, y_test) * 100
```

### Predict the data

```python
lr.predict([[5.3]])
```

```python
# Prediction result
y_pred_test = lr.predict(X_test)     # predicted value of y_test
y_pred_train = lr.predict(X_train)   # predicted value of y_train
```

## Step 7 : Plot the training and test results

### 1) Plot training set data vs predictions

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred_train)
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.show()
```

### 2) Plot test set data vs predictions

```python
plt.scatter(X_test, y_test)
plt.plot(X_train, y_pred_train)
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.show()
```

### we can plot results with the predicted value of y_test (lr.predict(X_test)) but the regression line would remain the same at it is generated from the unique equation of linear regression with the same training data.

## Coefficient and Intercept

```python
# Regressor coefficients and intercept

print(f'Coefficient: {lr.coef_}')
print(f'Intercept: {lr.intercept_}')
```
